# Extracting the tar file content into a byte array

```python
def extract_tar(filename):
    tar = open(filename, "rb").read()
    return tar
```

```python
sample1_tar = extract_tar("./data/sample1.tar")
print(len(sample1_tar) /(1024 * 1024))
```

```python
sample2_tar = extract_tar("./data/sample2.tar")
print(len(sample2_tar)/(1024 * 1024))
```

```python
def get_hashlist(tar, block_size):
    i=0
    hash_list = []
    while i < len(tar):
        start_index = i
        end_index = i + block_size
        if(end_index> len(tar)):
            end_index = len(tar)

        key = hash(tar[start_index : end_index])
        hash_list.append(key)
        i = i + block_size
    return hash_list
```

```python
len(get_hashlist(sample1_tar, 512)) * 512 /(1024 *1024)
```

```python
len(Counter(get_hashlist(sample1_tar, 512))) * 512 /(1024 *1024)
```

# Count instersection of all blocks which are available common in tar 1 and tar 2

```python
def get_intersection(tar1, tar2, block_size):
    hashlist1 = get_hashlist(tar1, block_size) 
    hashlist2 = get_hashlist(tar2, block_size) 
    return (sum((Counter(hashlist1) & Counter(hashlist2)).values()) * block_size) /(1024 * 1024)
```

# Returns all blocks in tar 2 which are also available in tar 1

```python
def get_intersection_method2(tar1, tar2, block_size):
    set1 = set(get_hashlist(tar1, block_size))
    hashlist2 = get_hashlist(tar2, block_size) 
    match_string =""
    match_size = 0
    for block in hashlist2:
        if block in set1:
            match_size += 1
            match_string += '1'
        else:
            match_string += '0'

    print(match_string)
    return (match_size * block_size)/(1024 * 1024)
```

```python
def read_in_chunks(file_object, chunk_size=512):
    """Lazy function (generator) to read a file piece by piece.
    Default chunk size: 512."""
    while True:
        data = file_object.read(chunk_size)
        if not data:
            break
        yield data

hash_list = []
with open('./top200.tar', 'rb') as f:
    for piece in read_in_chunks(f):
        key = hash(piece)
        hash_list.append(key)
```
